```
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

p=0.5
N=400
k=0.001
dt=0.01
n0=[0]*N
n1=[0]*N
n2=[0]*N
n3=[0]*N
n4=[0]*N
n0[0]=9996
n1[0]=1
n2[0]=1
n3[0]=1
n4[0]=1
for i in range(len(n0)-1):
  if(n0[i]<=0):
    n0[i+1]=0
    n0[i]=0
  if(n1[i]<=0):
    n1[i+1]=0
    n1[i]=0
  if(n2[i]<=0):
    n2[i+1]=0
    n2[i]=0
  if(n3[i]<=0):
    n3[i+1]=0
    n3[i]=0
  if(n4[i]<=0):
    n4[i+1]=0
    n4[i]=0
  n0[i+1]=n0[i]+(-p*k*n1[i]*n0[i]-p*k*n2[i]*n0[i]-p*k*n3[i]*n0[i]-p*k*n4[i]*n0[i])*dt
  n1[i+1]=n1[i]+(p*k*n0[i]*n1[i]-p*k*n2[i]*n1[i]-p*k*n3[i]*n1[i]-p*k*n4[i]*n1[i])*dt
  n2[i+1]=n2[i]+(p*k*n0[i]*n2[i]+p*k*n1[i]*n2[i]-p*k*n3[i]*n2[i]-p*k*n4[i]*n2[i])*dt
  n3[i+1]=n3[i]+(p*k*n0[i]*n3[i]+p*k*n1[i]*n3[i]+p*k*n2[i]*n3[i]-p*k*n4[i]*n3[i])*dt
  n4[i+1]=n4[i]+(+p*k*n0[i]*n4[i]+p*k*n1[i]*n4[i]+p*k*n2[i]*n4[i]+p*k*n3[i]*n4[i])*dt
plt.plot(n0,'red',label='Neutral voters')
plt.plot(n1,'blue',label='Candidate 1 voters')
plt.plot(n2,'green',label='Candidate 2 voters')
plt.plot(n3,'black',label='Candidate 3 voters')
plt.plot(n4,'orange',label='Candidate 4 voters')
plt.legend()
plt.title('Variation of number of voters with respect to time')
plt.xlabel('Time')
plt.ylabel('No of voters')
plt.show()
```
