```
# Cellular automata - ver 2
# Considering Moore neighbourhood
# uniform probability of conversion

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

N = 50

a = np.zeros((N,N),int)
b = np.zeros((N,N),int)

a[int(N/4)][int(N/4)] = 1
a[int(N/4)][int(3*N/4)] = 2
a[int(3*N/4)][int(N/4)] = 3
a[int(3*N/4)][int(3*N/4)] = 4

# for i in range(N):
#   for j in range(N):
#     print (a[i][j],end=" ")
#   print()
# print()
p=0.25

time=100
c_matrix = [[0] * (time+1), [0] * (time+1), [0] * (time+1), [0] * (time+1), [0] * (time+1)]

for t in range(time+1):

  figure(num=10, figsize=(8, 6), dpi=90, facecolor='w', edgecolor='k')
  if (t)%1 == 0:
    plt.imshow(a[1:],cmap=plt.cm.get_cmap('Blues', 12))
    plt.colorbar()
    plt.title('at Time = %i' %t)
    plt.show()
  
  
  for i in range(N):
    for j in range(N):
      
      # left
      if a[i][j] < a[i][(j-1)%N]:
        n = np.random.uniform(0,1)
        if n < p:
            b[i][j] = a[i][(j-1)%N]
      
      # right
      if b[i][j] < a[i][(j+1)%N]:
        n = np.random.uniform(0,1)
        if n < p:
          b[i][j] = a[i][(j+1)%N]
      
      # top
      if b[i][j] < a[(i-1)%N][j]:
        n = np.random.uniform(0,1)
        if n < p:
          b[i][j] = a[(i-1)%N][j]

      # bottom
      if b[i][j] < a[(i+1)%N][j]:
        n = np.random.uniform(0,1)
        if n < p:
          b[i][j] = a[(i+1)%N][j]
          
      # top-left
      if b[i][j] < a[(i-1)%N][(j-1)%N]:
        n = np.random.uniform(0,1)
        if n < p:
            b[i][j] = a[i][(j-1)%N]
      
      # top-right
      if b[i][j] < a[(i-1)%N][(j+1)%N]:
        n = np.random.uniform(0,1)
        if n < p:
          b[i][j] = a[i][(j+1)%N]
      
      # bottom-left
      if b[i][j] < a[(i+1)%N][(j-1)%N]:
        n = np.random.uniform(0,1)
        if n < p:
          b[i][j] = a[(i-1)%N][j]

      # bottom-right
      if b[i][j] < a[(i+1)%N][(j+1)%N]:
        n = np.random.uniform(0,1)
        if n < p:
          b[i][j] = a[(i+1)%N][j]
          
  for i in range(N):
    for j in range(N):
      if b[i][j] != 0:
          a[i][j] = b[i][j]     

  a[int(N/4)][int(N/4)] = 1
  a[int(N/4)][int(3*N/4)] = 2
  a[int(3*N/4)][int(N/4)] = 3
  a[int(3*N/4)][int(3*N/4)] = 4
  
  for i in range(N):
    for j in range(N):
      c_matrix[a[i][j]][t] = c_matrix[a[i][j]][t] + 1
      
figure(num=10, figsize=(8, 6), dpi=90, facecolor='w', edgecolor='k')
plt.plot(c_matrix[0])
plt.plot(c_matrix[1])
plt.plot(c_matrix[2])
plt.plot(c_matrix[3])
plt.plot(c_matrix[4])
plt.legend(['No opinion','Party 1', 'Party 2', 'Party 3', 'Party 4'])
plt.title('Party size')
plt.xlabel('Time (days)')
plt.ylabel('Number of people')

#   for i in range(N):
#     for j in range(N):
#       print (a[i][j],end=" ")
#     print()
#   print()
```

```
# Cellular automata - ver 3
# Considering Moore neighbourhood
# non-uniform probability of conversion

import numpy as np

N = 50
num_can = 4
a = np.zeros((N,N),int)
b = np.zeros((N,N),int)

a[int(N/4)][int(N/4)] = 1
a[int(N/4)][int(3*N/4)] = 2
a[int(3*N/4)][int(N/4)] = 3
a[int(3*N/4)][int(3*N/4)] = 4

# for i in range(N):
#   for j in range(N):
#     print (a[i][j],end=" ")
#   print()
# print()
p=0.5

time=100
c_matrix = [[0] * (time+1), [0] * (time+1), [0] * (time+1), [0] * (time+1), [0] * (time+1)]

for t in range(time+1):
  
  figure(num=10, figsize=(8, 6), dpi=90, facecolor='w', edgecolor='k')
  if (t)%25 == 0:
    plt.imshow(a[1:],cmap=plt.cm.get_cmap('Blues', 12))
    plt.colorbar()
    plt.title('at Time = %i' %t)
    plt.show()
  
  for i in range(N):
    for j in range(N):
      
      # left
      if a[i][j] < a[i][(j-1)%N]:
        n = np.random.uniform(0,1)
        if n < (0.3 + (0.2)/(num_can - a[i][(j-1)%N] + 1)):
            b[i][j] = a[i][(j-1)%N]
      
      # right
      if b[i][j] < a[i][(j+1)%N]:
        n = np.random.uniform(0,1)
        if n < (0.3 + 0.2/(num_can - a[i][(j+1)%N] + 1)):
          b[i][j] = a[i][(j+1)%N]
      
      # top
      if b[i][j] < a[(i-1)%N][j]:
        n = np.random.uniform(0,1)
        if n < (0.3 + (0.2)/(num_can - a[(i-1)%N][j] + 1)):
          b[i][j] = a[(i-1)%N][j]

      # bottom
      if b[i][j] < a[(i+1)%N][j]:
        n = np.random.uniform(0,1)
        if n < (0.3 + (0.2)/(num_can - a[(i+1)%N][j] + 1)):
          b[i][j] = a[(i+1)%N][j]
          
      # top-left
      if b[i][j] < a[(i-1)%N][(j-1)%N]:
        n = np.random.uniform(0,1)
        if n < (0.3 + (0.2)/(num_can - a[(i-1)%N][(j-1)%N] + 1)):
            b[i][j] = a[i][(j-1)%N]
      
      # top-right
      if b[i][j] < a[(i-1)%N][(j+1)%N]:
        n = np.random.uniform(0,1)
        if n < (0.3 + (0.2)/(num_can - a[(i-1)%N][(j+1)%N] + 1)):
          b[i][j] = a[i][(j+1)%N]
      
      # bottom-left
      if b[i][j] < a[(i+1)%N][(j-1)%N]:
        n = np.random.uniform(0,1)
        if n < (0.3 + (0.2)/(num_can - a[(i+1)%N][(j-1)%N] + 1)):
          b[i][j] = a[(i-1)%N][j]

      # bottom-right
      if b[i][j] < a[(i+1)%N][(j+1)%N]:
        n = np.random.uniform(0,1)
        if n < (0.3 + (0.2)/(num_can - a[(i+1)%N][(j+1)%N] + 1)):
          b[i][j] = a[(i+1)%N][j]
          
  for i in range(N):
    for j in range(N):
      if b[i][j] != 0:
          a[i][j] = b[i][j]     

  a[int(N/4)][int(N/4)] = 1
  a[int(N/4)][int(3*N/4)] = 2
  a[int(3*N/4)][int(N/4)] = 3
  a[int(3*N/4)][int(3*N/4)] = 4
  
  for i in range(N):
    for j in range(N):
      c_matrix[a[i][j]][t] = c_matrix[a[i][j]][t] + 1
  
  
figure(num=10, figsize=(8, 6), dpi=90, facecolor='w', edgecolor='k')
plt.plot(c_matrix[0])
plt.plot(c_matrix[1])
plt.plot(c_matrix[2])
plt.plot(c_matrix[3])
plt.plot(c_matrix[4])
plt.legend(['No opinion','Party 1', 'Party 2', 'Party 3', 'Party 4'])
plt.title('Party size')
plt.xlabel('Time (days)')
plt.ylabel('Number of people')
      
#   for i in range(N):
#     for j in range(N):
#       print (a[i][j],end=" ")
#     print()
#   print()
```

```
# Cellular Automata - ver 4
# majority has power not the candidate themselves

import numpy as np
import matplotlib.pyplot as plt


from matplotlib.pyplot import figure

# np.random.seed(34)

N = 50
num_can = 4
a = np.zeros((N,N),int)
b = np.zeros((N,N),int)

a[int(N/4)][int(N/4)] = 1
a[int(N/4)][int(3*N/4)] = 2
a[int(3*N/4)][int(N/4)] = 3
a[int(3*N/4)][int(3*N/4)] = 4
 

# for i in range(N):
#   for j in range(N):
#     print (a[i][j],end=" ")
#   print()
# print()
p=0.5

time=100

c_matrix = [[0] * (time+1), [0] * (time+1), [0] * (time+1), [0] * (time+1), [0] * (time+1)]

for t in range(time+1):
  
  # the color bar function
  figure(num=10, figsize=(8, 6), dpi=90, facecolor='w', edgecolor='k')
  if (t)%1 == 0:
    plt.imshow(a[1:],cmap=plt.cm.get_cmap('Blues', 12))
    plt.colorbar()
    plt.title('at Time = %i' %t)
    plt.show()

  for i in range(N):
    for j in range(N):
      count = [0] * (num_can+1)
      
      if a[(i-1)%N][(j-1)%N] != 0:
        count[a[(i-1)%N][(j-1)%N]] = count[a[(i-1)%N][(j-1)%N]] + 1
      
      if a[(i-1)%N][j] != 0:
        count[a[(i-1)%N][j]] = count[a[(i-1)%N][j]] + 1

      if a[(i-1)%N][(j+1)%N] != 0:
        count[a[(i-1)%N][(j+1)%N]] = count[a[(i-1)%N][(j+1)%N]] + 1

      if a[i][(j-1)%N] != 0:
        count[a[i][(j-1)%N]] = count[a[i][(j-1)%N]] + 1

      if a[i][(j+1)%N] != 0:
        count[a[i][(j+1)%N]] = count[a[i][(j+1)%N]] + 1 

      if a[(i+1)%N][(j-1)%N] != 0:
        count[a[(i+1)%N][(j-1)%N]] = count[a[(i+1)%N][(j-1)%N]] + 1
      
      if a[(i+1)%N][j] != 0:
        count[a[(i+1)%N][j]] = count[a[(i+1)%N][j]] + 1 
      
      if a[(i+1)%N][(j+1)%N] != 0:
        count[a[(i+1)%N][(j+1)%N]] = count[a[(i+1)%N][(j+1)%N]] + 1
      
      c_max = max(count)
      if count.index(c_max) != 0:
        if c_max >= 4:
          n = np.random.uniform(0,1)
          if n < 0.5:
            b[i][j] = count.index(c_max)
        else:
          n = np.random.uniform(0,1)
          if n < c_max/10:
            b[i][j] = count.index(c_max)
     
  for i in range(N):
    for j in range(N):
        a[i][j] = b[i][j]     

  a[int(N/4)][int(N/4)] = 1
  a[int(N/4)][int(3*N/4)] = 2
  a[int(3*N/4)][int(N/4)] = 3
  a[int(3*N/4)][int(3*N/4)] = 4
  
  for i in range(N):
    for j in range(N):
      c_matrix[a[i][j]][t] = c_matrix[a[i][j]][t] + 1
  

figure(num=10, figsize=(8, 6), dpi=90, facecolor='w', edgecolor='k')
plt.plot(c_matrix[0])
plt.plot(c_matrix[1])
plt.plot(c_matrix[2])
plt.plot(c_matrix[3])
plt.plot(c_matrix[4])
plt.legend(['No opinion','Party 1', 'Party 2', 'Party 3', 'Party 4'])
plt.title('Party size')
plt.xlabel('Time (days)')
plt.ylabel('Number of people')
      
# for i in range(N):
#   for j in range(N):
#     print (a[i][j],end=" ")
#   print()
# print()

```

```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from mpl_toolkits.axes_grid1 import make_axes_locatable

fig = plt.figure()
ax = fig.add_subplot(111)

# I like to position my colorbars this way, but you don't have to
div = make_axes_locatable(ax)
cax = div.append_axes('right', '5%', '5%')

def f(x, y):
    return np.exp(x) + np.sin(y)

x = np.linspace(0, 1, 120)
y = np.linspace(0, 2 * np.pi, 100).reshape(-1, 1)

# This is now a list of arrays rather than a list of artists
frames = []
for i in range(10):
    x       += 1
    curVals  = f(x, y)
    frames.append(curVals)

cv0 = frames[0]
im = ax.imshow(cv0, origin='lower') # Here make an AxesImage rather than contour
cb = fig.colorbar(im, cax=cax)
tx = ax.set_title('Frame 0')

def animate(i):
    arr = frames[i]
    vmax     = np.max(arr)
    vmin     = np.min(arr)
    im.set_data(arr)
    im.set_clim(vmin, vmax)
    tx.set_text('Frame {0}'.format(i))
    # In this version you don't have to do anything to the colorbar,
    # it updates itself when the mappable it watches (im) changes

ani = animation.FuncAnimation(fig, animate, frames=10)

plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
# plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(5, 3))
ax.set(xlim=(-3, 3), ylim=(-1, 1))

x = np.linspace(-3, 3, 91)
t = np.linspace(1, 25, 30)
X2, T2 = np.meshgrid(x, t)
 
sinT2 = np.sin(2*np.pi*T2/T2.max())
F = 0.9*sinT2*np.sinc(X2*(1 + sinT2))

line = ax.plot(x, F[0, :], color='k', lw=2)[0]

def animate(i):
    line.set_ydata(F[i, :])
anim = FuncAnimation(
    fig, animate, interval=100, frames=len(t)-1)
 
plt.draw()
plt.show()
```
